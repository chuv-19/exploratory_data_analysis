7. Выведите на экран медиану и среднее арифметическое столбцов:
 - Rating
 - Reviews

```python
def print_scores(df, column):
    print(f'Среднее значение колонки {column} = {df[column].mean()}')
    print(f'Медианное значение колонки {column} = {df[column].median()}')


#Дропнула строки, в которых не могла перевести значение в int
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df = df.dropna(subset=['Reviews'])

print_scores(df, 'Rating')
print_scores(df, 'Reviews')
```

8. Выведите на экран все уникальные значения категориального столбца Genres

```python
#Если имеется ввиду уникальные значения жанров, а не их комбинаций
uniq_g = set()
for i in df['Genres'].values:
    for j in i.split(';'):
        uniq_g.add(j)
print(uniq_g)

#Если имеется ввиду комбинации
print(df['Genres'].unique())
```

9. Сгруппируйте данные по столбцу `Genres` и посчитайте для каждого жанра средний и медианный рейтинг. Представьте результат в виде нового DataFrame,который будет сохранен в переменную `grouped_df`, где одна колонка будет содержать названия жарнов, а две другие буду содержать в себе средний и медианный рейтинги. Напишите краткий вывод.

```python
grouped_df = df.groupby('Genres').agg({'Rating': ['mean', 'median']})
grouped_df.columns = ['mean', 'median']
grouped_df

#Использовали groupby, чтобы по нему сгруппировать жанры и agg, чтобы использовать нужные нам функции дня нового df
```

10. Из получившегося датафрейма выведите на экран только те жанры, чей медианный рейтинг больше 4.5. Напишите краткий вывод.

```python
grouped_df[grouped_df['median'] > 4.5]
#Базовый элемент работы с df в pd
```
